```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings;
warnings.filterwarnings('ignore')
%matplotlib inline 
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+13, color="green", lw=3, ls='-', marker='o', markersize=8)
```

```python
x = np.linspace(0, 5, 11)
fig, ax = plt.subplots(figsize=(6,3))
ax.plot(x, x+13, color="green", lw=3, ls='-', marker='o', markersize=10)
ax.set_xlim([0, 3])
ax.set_ylim([13, 16]);
```

```python
x = np.linspace(0, 5, 11)
fig, axes = plt.subplots()    
axes.plot(x, x**2, label="square")
axes.plot(x, x**3, label="cube")
axes.legend()
fig.savefig("square_cube.png")
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]

salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
x = np.arange(0, 11)
y = x.copy()
z = x*2
t = np.log(x)
```

```python
fig, ax= plt.subplots()
ax.plot(x,y, "r")
ax.plot(x,z, "b")
ax.plot(x,t, "k")
ax.set_xlabel("x")
ax.set_ylabel("y or z or t")
ax.set_title("Title");

plt.show()
```

```python
fig = plt.figure()      # ax = fig.add_axes([left, bottom, widht, height]) 
                                          # left, bottom, width, height (range 0 to 1)
ax1 = fig.add_axes([0.0, 0.0, 1, 1])
ax2 = fig.add_axes([0.7, 0.1, 0.3, 0.3])

#fig, ax = plt.subplots(2,1)
#ax[0].plot(age, salary, "r")
#ax[1].plot(age, salary_2);

ax1.plot(age, salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")

ax2.plot(age, salary_2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(age,salary, 'r')
ax[0,1].plot(age,salary_2, 'y')
ax[1,0].plot(age,salary, 'g')
ax[1,1].plot(age,salary_2, 'b')
plt.tight_layout();
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
# Use marker for string code
# Use markersize or ms for size
ax.plot(x, x-1, marker='+', markersize=20)
ax.plot(x, x-2, marker='o', ms=20) #ms can be used for markersize
ax.plot(x, x-3, marker='s', ms=20, lw=0) # make linewidth zero to see only markers
ax.plot(x, x-4, marker='1', ms=20)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.suptitle("LineStyles & Markers & Markersize & Color", fontsize=16, y = 1.2)
plt.subplot(2, 2, 1)
plt.plot(age, salary, "r", ls="-", marker="+" , markersize=30)
plt.subplot(2, 2, 2)
plt.plot(age, salary_2, "b", ls="-.", marker="o" , markersize=30)
plt.subplot(2, 2, 3)
plt.plot(age, salary, "g", ls=":", marker="x" , markersize=30)
plt.subplot(2, 2, 4)
plt.plot(age, salary_2, "y", ls="--" , marker="*" , markersize=30)
plt.tight_layout();
plt.show()
```

```python
tips = sns.load_dataset("tips")
```

```python
plt.figure(figsize= (8,6)) # figure boyutu bytlebilir
plt.scatter(tips["tip"], tips["total_bill"])
```

```python
day = tips.groupby("day").sum().index
day
```

```python
day_of_total_bill_list = list(tips.groupby("day")["total_bill"].sum())
day_of_total_bill_list
```

```python
fig, ax = plt.subplots()
ax.bar(day, day_of_total_bill_list)
plt.show()
```

```python
day_tip = np.array(tips.groupby("day")["tip"].sum())
day_tip
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(day, day_of_total_bill_list,  label="total_bill")
ax.bar(day, day_tip,            label="tip")
plt.legend();
```
